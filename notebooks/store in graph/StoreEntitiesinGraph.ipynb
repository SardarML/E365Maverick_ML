## Globals and Imports

```python
from os.path import isfile, join
from os import listdir
from pathlib import Path
import sys
from rdflib import RDF, SDO, Graph, URIRef, BNode
import logging
import json
import requests
```

```python
# for loading the videos

base = "D:\\Scraped\\youtube\\rdf\\video"
host = "https://entitygraph.azurewebsites.net/"
path = "/api/entities"
key = "Tzre7295T10z1K"
```

## Functions

```python
def loadFiles(): 
   return [f for f in listdir(base) if isfile(join(base, f))]

def openFile(filename: str) -> str: 
    file: Path =  Path.joinpath(Path(base), filename)
    content = file.open(mode='r', encoding="utf-8").read()
    return content

def storeContent(content: str) -> bool: 
    url = base+path
    body = content.encode("utf-8")
    try: 
        response = requests.post(
            url=host+path, 
            data=body, 
            headers= {
                "X-API-KEY": key, 
                "Content-Type": "text/turtle"
            })
        
        if not response.ok: 
            print("Request failed with message: "+response.reason)
    except Exception as err:
        print("failed to store. reason: "+str(err))
```

## Script

```python
file = loadFiles()[0]
print(file)
data: str = openFile(filename=file)
storeContent(content=data)
```

```python
for file in loadFiles(): 
    print(file)
    data: str = openFile(filename=file)
    storeContent(content=data)
    
```
