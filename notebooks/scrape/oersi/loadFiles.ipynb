## Search example
Showing how to access the API

```python
import json
import requests
# curl -X 'POST' 'https://oersi.org/resources/api/search/oer_data/_search?pretty' -H 'Content-Type: application/json' -H 'accept: application/json' -d '{"size":20,"from":0,"query": { "multi_match": { "query": "Klimawandel", "fields": ["name", "description", "keywords"]}},"sort": [{"id":"asc"}]}'
body = {
        "size": 20, 
        "from": 0, 
        "query": {
            "multi_match": {
                "query": "Klimawandel",
                "fields": [ "name", "description", "keywords" ]
            }
        }, 
        "sort": [ { "id": "asc" }]
}


    

response = requests.post(
    url="https://oersi.org/resources/api/search/oer_data/_search?pretty", 
    data= json.dumps(body), 
    headers= {
        "accept": "application/json", 
        "Content-Type": "application/json"
    })
res = json.loads(response.content)

for hit in res["hits"]["hits"]: 
    print(json.dumps(hit["_source"], indent=4))
```
