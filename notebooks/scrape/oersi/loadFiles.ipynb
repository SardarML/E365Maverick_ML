## Search example
Showing how to access the API

```python
import json
import requests
# curl -X 'POST' 'https://oersi.org/resources/api/search/oer_data/_search?pretty' -H 'Content-Type: application/json' -H 'accept: application/json' -d '{"size":20,"from":0,"query": { "multi_match": { "query": "Klimawandel", "fields": ["name", "description", "keywords"]}},"sort": [{"id":"asc"}]}'
body = {
        "size": 20, 
        "from": 0, 
        "query": {
            "multi_match": {
                "query": "Klimawandel",
                "fields": [ "name", "description", "keywords" ]
            }
        }, 
        "sort": [ { "id": "asc" }]
}


    

response = requests.post(
    url="https://oersi.org/resources/api/search/oer_data/_search?pretty", 
    data= json.dumps(body), 
    headers= {
        "accept": "application/json", 
        "Content-Type": "application/json"
    })
res = json.loads(response.content)

for hit in res["hits"]["hits"]: 
    print(json.dumps(hit["_source"], indent=4))
```

# Creating the Pit Id

```python
import requests

headers = {
    'accept': 'application/json',
}

response = requests.post('https://oersi.org/resources/api/search/oer_data/_pit?keep_alive=15m&pretty', headers=headers)
```

```python
res = json.loads(response.content)
```

```python
print(res['id'])
```

```python
pit_id = res['id']
```

# Searching and Showing first 500 results

```python
headers = {
    'Content-Type': 'application/json',
    'accept': 'application/json',
}

json_data = {
    'size': 500,
    'query': {
        'match': {
            'mainEntityOfPage.provider.name': 'twillo',
        },
    },
    'pit': {
        'id': pit_id,
        'keep_alive': '15m',
    },
    'sort': [
        {
            'id': 'asc',
        },
    ],
    'track_total_hits': True,
}

response = requests.post('https://oersi.org/resources/api/search/_search?pretty', headers=headers, json=json_data)
```

```python
res = json.loads(response.content)
print(res)
```

## Determine YOUR_LAST_SORT_RESULT (sort entry of the last hit)

```python
last_sort_result = res['hits']['hits'][-1]['sort']
```

## Repeat search until no more hits are found

```python
while res['hits']['hits']:
    last_sort_result = res['hits']['hits'][-1]['sort']
    
    headers = {
    'Content-Type': 'application/json',
    'accept': 'application/json',
    }

    json_data = {
    'size': 500,
    'query': {
        'match': {
            'mainEntityOfPage.provider.name': 'twillo',
        },
    },
    'pit': {
        'id': pit_id,
        'keep_alive': '15m',
    },
    'sort': [
        {
            'id': 'asc',
        },
    ],
    'track_total_hits': True,
    'search_after': last_sort_result
    }

    response = requests.post('https://oersi.org/resources/api/search/_search?pretty', headers=headers, json=json_data)
    res = json.loads(response.content)
    print(len(res['hits']['hits']))
    print(last_sort_result)
     
```

## Delete the Pit

```python
# delete the pit

headers = {
    'Content-Type': 'application/json',
}

json_data = {
    'id': pit_id,
}

response = requests.delete('https://oersi.org/resources/api/search/_pit', headers=headers, json=json_data)
```

```python
res = json.loads(response.content)
print(res)
```
